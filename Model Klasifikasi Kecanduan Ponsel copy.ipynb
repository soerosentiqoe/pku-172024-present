```python
import pandas as pd 
```

```python
dataPonsel = pd.read_csv('dataponselnew.csv')
```

```python
data = dataPonsel.copy()
```

```python
target = 'kecanduan'
```

```python
encode = ['gender', 'jenjangPekerjaan','intensitasSosmed','gangguanTidur','cemasPonselTertinggal','gangguanProduktivitas']
```

```python
data
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'Ya':0, 'Tidak':1}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['kecanduan'] = data['kecanduan'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('kecanduan', axis=1)
Y = data['kecanduan']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<h3>SVM</h3>

```python
from sklearn.svm import SVC
import numpy as np
from sklearn.metrics import classification_report
#y_train_2 = np.argmax(y_train, 1)
svmClf = SVC(kernel='linear')
svmClf.fit(X_train, y_train)
y_preds = svmClf.predict(X_test)
print(classification_report(y_test, y_preds))
```

<h3>Decision Tree<h3>

```python
from sklearn.tree import DecisionTreeClassifier
dtreeClf = DecisionTreeClassifier(criterion="gini")
dtreeClf.fit(X_train, y_train)
y_preds = dtreeClf.predict(X_test)
print(classification_report(y_test, y_preds))
```

<h3>Xgboost</h3>

```python
from xgboost import XGBClassifier

xgbClf = XGBClassifier()
xgbClf.fit(X_train, y_train)
y_preds = xgbClf.predict(X_test)
print(classification_report(y_test, y_preds))
```

<h3>Random Forest</h3>

```python
from sklearn.ensemble import RandomForestClassifier

rfClf = RandomForestClassifier(n_estimators=50)
rfClf.fit(X_train, y_train)
y_preds = rfClf.predict(X_test)
print(classification_report(y_test, y_preds))
```
